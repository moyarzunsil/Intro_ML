# Regresión Lineal

En este tutorial vamos a calcular el predictor más sencillo de todos, la __Regresión Lineal__. Para ello usaremos las librerías _numpy_ y _sklearn_. Pese a que esta función se encuentra implementada en la mayoría de las librerías matemáticas, hoy la calcularemos a través del _Método de Mínimos Cuadrados Ordinarios_.

Hablamos de una __Regresión Lineal__ según [Wikipedia](https://es.wikipedia.org/wiki/Regresi%C3%B3n_lineal):

_"En estadística la regresión lineal o ajuste lineal es un modelo matemático usado para aproximar la relación de dependencia entre una variable dependiente $Y$, las variables independientes $X_i$ y un término aleatorio $\epsilon$. Este modelo puede ser expresado como: $$Y_t=\beta_0+\beta_1X_1+\beta_2X_2+\ldots+\beta_pX_p+\epsilon$$_
_donde:_

_$Y_t$: variable dependiente, explicada o regresando._

_$X_1,X_2,\ldots,X_p$: variables explicativas, independientes o regresores._

_$\beta_0,\beta_1,\beta_2,\ldots,\beta_p$: parámetros, miden la influencia que las variables explicativas tienen sobre el regrediendo._

_donde $\beta_0$ es la intersección o término "constante", las $\beta_i (i>0)$ son los parámetros respectivos a cada variable independiente, y $p$ es el número de parámetros independientes a tener en cuenta en la regresión."_

```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
```

```python
boston=load_boston()

print(boston.DESCR)
```

```python
X=np.array(boston.data[:,5])
Y=np.array(boston.target)

plt.scatter(X,Y,alpha=0.3)
plt.show()
```

Resolvemos a través de los Mínimos Cuadrados Ordinarios, según [Wikipedia](https://es.wikipedia.org/wiki/M%C3%ADnimos_cuadrados_ordinarios):

_"En estadística, los mínimos cuadrados ordinarios (MCO) o mínimos cuadrados lineales es el nombre de un método para encontrar los parámetros poblacionales en un modelo de regresión lineal. Este método minimiza la suma de las distancias verticales entre las respuestas observadas en la muestra y las respuestas del modelo. El parámetro resultante puede expresarse a través de una fórmula sencilla, especialmente en el caso de un único regresionador."_

Cuya fórmula matricial es: $$\hat{\beta}=(X^TX)^{-1}X^TY$$

```python
X=np.array(boston.data[:,5])
Y=np.array(boston.target)

X=np.array([np.ones(X.size),X]).T

B=np.linalg.inv(X.T@X)@X.T@Y

print(B)
```

```python
plt.scatter(X[:,1],Y,alpha=0.3)
plt.plot([4,9], [B[0]+B[1]*4,B[0]+B[1]*9],c="red")
plt.show()
```
